## Модель

### Подготовка выборок

Выделим целевые признаки, которых не хватает в тестовой выборке.

```python
target = data_full[['date', 'rougher.output.recovery', 'final.output.recovery']]
target.head(10)
```

Подготовим выборки. Добавим в тестовую целевые признаки без даты, а из тренировочной удалим лишние признаки.

```python
data_test = data_test.merge(target, how = 'left', on = 'date')
data_test.drop(['date'], axis=1, inplace=True)
data_train = data_train.drop(set(data_train.columns)-set(data_test.columns), axis=1)
```

Напишем функцию для вычисления sMAPE.

```python
def smape(y, y_pred):
    return (((abs(y - y_pred) * 2 * 100) /
              (abs(y) + abs(y_pred))).sum()) / len(y)

def end_smape(rougher, final):
    return 0.25 * rougher.item() + 0.75 * final.item()
```

### Обучение и оценка модели

Финальная метрика end_smape линейно зависит от smape после флотации и после очистки. Выделим в обучающей выборке атрибуты и целевые признаки

```python
# для концентрата сырья
features_train_r = data_train.drop(['rougher.output.recovery'], axis=1)
target_train_r = data_train['rougher.output.recovery']

# для финального концентрата
features_train_f = data_train.drop(['rougher.output.recovery', 'final.output.recovery'], axis=1)
target_train_f = data_train['final.output.recovery']

# для чернового концентрата сырья
features_test_r = data_test.drop(['rougher.output.recovery'], axis=1)
target_test_r = data_test['rougher.output.recovery']

# для финального концентрата 
features_test_f = data_test.drop(['rougher.output.recovery', 'final.output.recovery'], axis=1)
target_test_f = data_test['final.output.recovery']
```

Убедимся, что обучающие и тестовые выборки содержат одинаковые признаки.

Настроим гиперпараметры.

```python
def get_cv_scores(model, target, features):
    my_scorer = make_scorer(smape)
    scores = cross_val_score(model, features,target, cv=5, scoring=my_scorer)
    return scores.mean()
```

#### Древо решений

```python
from collections import defaultdict
d = defaultdict(list)

for depth in range(2, 14):
    for leaf in range(1, 7):
        model_1 = DecisionTreeRegressor(max_depth=depth, min_samples_leaf = leaf, random_state=12345)
        model_2 = DecisionTreeRegressor(max_depth=depth, min_samples_leaf = leaf, random_state=12345)
        rougher_smape = get_cv_scores(model_1, target_train_r, features_train_r)
        final_smape = get_cv_scores(model_2, target_train_f, features_train_f)
        d['rougher_smape'].append(rougher_smape)
        d['final_smape'].append(final_smape)
        d['end_smape'].append(end_smape(rougher_smape, final_smape))
        d['depth'].append(depth)
        d['leaf'].append(leaf)

pd.DataFrame(d).sort_values(by='end_smape')
```

Из данной модели наиболее лучший sMAPE при следующих параметрах: min_sample leaf - 1, max_depth - 2, end_smape - 8.249558, final_smape - 8.918837, rougher_smap - 6.241723.
